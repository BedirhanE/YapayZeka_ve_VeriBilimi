```python
import seaborn as sns   
from pandas.api.types import CategoricalDtype
diamonda = sns.load_dataset('diamonds')
df= diamonda.copy()
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
df.head()
```

```python

sns.catplot(x = "cut", y = "price", hue = "color", kind = "point", data = df);# iki kategorik değişkenin ilişkisini göstermek için kullanılır.

```

```python
sns.barplot(x = "cut", y = "price", hue = "color", data = df); # barplot ile kategorik değişkenlerin ortalamalarını görebiliriz.
#daha fazla açıklanabilirlik elde etmiş olduk bu yöntemelri kullanarak.
```

```python
import seaborn as sns
diamonda = sns.load_dataset('diamonds')
df= diamonda.copy()
df.head()
```

```python
sns.kdeplot(df.price,shade=True);#iki değişkenin yoğunluk grafiğini gösterir.
```

```python
sns.FacetGrid(df, hue = "cut", height = 5).map(sns.kdeplot, "price",shade=True).add_legend();# FacetGrid ile iki değişkenin yoğunluk grafiğini gösterir.
```

```python
sns.catplot(x = "cut", y = "price", hue = "color",kind = "point", data = df);# catplot ile iki değişkenin ilişkisini gösterir.
```
